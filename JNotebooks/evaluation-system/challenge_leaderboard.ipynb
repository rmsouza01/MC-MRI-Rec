```python
%matplotlib inline
import numpy as np
import os
import glob
import pickle
import h5py
import sys
from mrrec.metrics import metrics

results = glob.glob("../../../Results-submissions/Metrics/*.pickle")
nslices = 156
nfiles = 50

print("Track 01 - 12-channel")
print("R = 5")
t1_r5 = 'track01_12_channel_R=5'
#print(results)
for ii in results:
    name = ii.split("/")[-1].split(".pickle")[0]
    
    res_file = open(ii,'rb')
    res = pickle.load(res_file)
    res_file.close()
    
    if t1_r5 in res:
        res = res[t1_r5]
        ssim = res[:,0].reshape(nslices,nfiles)
        psnr = res[:,1].reshape(nslices,nfiles)
        vif = res[:,2].reshape(nslices,nfiles)
        print("%s &  $%.4f \\pm %.4f$ & $%.4f \\pm %.4f$ & $%.4f \pm %.4f$" %(name,ssim.mean(),ssim.std(),psnr.mean(),psnr.std(),\
                                                                  vif.mean(),vif.std()))

print("\n\n")

print("R = 10")
t1_r10 = 'track01_12_channel_R=10'
for ii in results:
    name = ii.split("/")[-1].split(".pickle")[0]
    
    res_file = open(ii,'rb')
    res = pickle.load(res_file)
    res_file.close()
    
    if t1_r10 in res:
        res = res[t1_r10]
        ssim = res[:,0].reshape(nslices,nfiles)
        psnr = res[:,1].reshape(nslices,nfiles)
        vif = res[:,2].reshape(nslices,nfiles)
        print("%s &  $%.4f \\pm %.4f$ & $%.4f \\pm %.4f$ & $%.4f \pm %.4f$" %(name,ssim.mean(),ssim.std(),psnr.mean(),psnr.std(),\
                                                                  vif.mean(),vif.std()))
```
