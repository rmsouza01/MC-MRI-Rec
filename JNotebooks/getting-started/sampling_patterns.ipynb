# Sampling Patterns

We provide sampling patterns for the different matrix sizes in the phase-encode and slice-encode directions. Remember that in the slice-encode direction only 85% of the k-space data was collected

- 218x170 - R = {5,10}
- 218x174 - R = {5,10}
- 218x178 - R = {5,10}

The sampling patterns follow a Poisson disc distribution and centre of k-space was were fully sampled withhin a radius of 16.

```python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
from mrrec.constants import DATA_PATH
```

## R = 5

```python
# Sampling masks, R = 5 for different matrix sizes
sp_r5 = [np.load(DATA_PATH / "poisson_sampling" / "R5_218x170.npy"),\
         np.load(DATA_PATH / "poisson_sampling" / "R5_218x174.npy"),\
         np.load(DATA_PATH / "poisson_sampling" / "R5_218x180.npy")]

sr = 0.85 # Sampling-rate in the slice-encode direction

plt.figure(figsize = (12,6))

for i,j in enumerate(sp_r5):
    plt.subplot(1,3,i + 1)
    plt.imshow(j[0],cmap = "gray")
    _,h,w = j.shape
    w2 = int(np.ceil(w*sr))
    plt.xticks([w2], [str(w2)], color='red')
    plt.title("Sampling:%.2f , size: %d x %d" %((j[0].sum()/j[0,:,:w2].size),\
                                                 j.shape[1],j.shape[2]))
plt.show()
```
